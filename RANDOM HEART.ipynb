# <font color = 'lime'> Selecionar dados randômico em um data set csv

```python
import numpy as np
import pandas as pd


df = pd.read_csv('heart.csv')
df
```

```python
df.shape
```

```python
df_index = list(df.index)
```

```python
indexs = np.random.choice(df_index,50)
```

```python
new_df = df.iloc[indexs]
```

```python
new_df.head()
```

```python
for i in new_df:
    print(i)
```

```python
df = new_df.reset_index()
df.head()
```

```python
df.columns.values
```

```python
new_df = df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
       'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']]
```

```python
for i in new_df:
    print(i)
```
